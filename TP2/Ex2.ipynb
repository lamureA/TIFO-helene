## Kernel Construction

```python
def gaussian_kernel(sigma, size):
    kernel = []
    for i in range(size):
        row = []
        for j in range(size):
            value = (1.0 / 2.0 * np.pi * sigma * sigma) * np.exp((-1.0 / 2.0 * sigma * sigma) * (i * i + j * j))
            row.append(value)
        kernel.append(row)
    return np.asarray(kernel)
```

```python
def mean_kernel(size):
    kernel = []
    for i in range(size):
        row = []
        for j in range(size):
            row.append(1.0 / size * size)
        kernel.append(row)
    return np.asarray(kernel)
```
