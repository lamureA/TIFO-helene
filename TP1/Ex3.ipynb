```python
from PIL import Image
import numpy as np
import matplotlib.pylab as plt
```

## Load images

```python
img1 = Image.open('../images/barbara.png')
img1.load()
data1 = np.asarray(img1)
```

```python
plt.imshow(data1, cmap='gray')
```

```python
img2 = Image.open('../images/lena.bmp')
img2.load()
data2 = np.asarray(img2)
```

```python
plt.imshow(data2, cmap='gray')
```

## FFT

```python
fft1 = np.fft.fft2(data1)
fft1 = np.fft.fftshift(fft1)
```

```python
fft2 = np.fft.fft2(data2)
fft2 = np.fft.fftshift(fft2)
```

## Combine

```python
fft_combine1 = abs(fft1) * np.exp(1j * np.angle(fft2))
ifft_combine1 = np.fft.ifftn(fft_combine1)
```

```python
plt.imshow(abs(ifft_combine1), cmap='gray')
```

```python
fft_combine2 = abs(fft2) * np.exp(1j * np.angle(fft1))
ifft_combine2 = np.fft.ifftn(fft_combine2)
plt.imshow(abs(ifft_combine2), cmap='gray')
```

## Compare modulus and phase

```python
fig=plt.figure(figsize=(10, 10))

fig.add_subplot(2, 2, 1)
plt.imshow(abs(np.log(fft1)), cmap='gray')

fig.add_subplot(2, 2, 2)
plt.imshow(abs(np.log(np.fft.fft2(ifft_combine1))), cmap='gray')

fig.add_subplot(2, 2, 3)
plt.imshow(np.angle(np.log(fft1)), cmap='gray')

fig.add_subplot(2, 2, 4)
plt.imshow(np.angle(np.log(np.fft.fft2(ifft_combine1))), cmap='gray')
```

```python
fig=plt.figure(figsize=(10, 10))

fig.add_subplot(2, 2, 1)
plt.imshow(abs(np.log(fft2)), cmap='gray')

fig.add_subplot(2, 2, 2)
plt.imshow(abs(np.log(np.fft.fft2(ifft_combine2))), cmap='gray')

fig.add_subplot(2, 2, 3)
plt.imshow(np.angle(np.log(fft2)), cmap='gray')

fig.add_subplot(2, 2, 4)
plt.imshow(np.angle(np.log(np.fft.fft2(ifft_combine2))), cmap='gray')
```
