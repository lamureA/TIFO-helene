```python
from PIL import Image
import numpy as np
import matplotlib.pylab as plt
```

## Load image

```python
img = Image.open('images/boat512.gif')
img.load()
data = np.asarray(img)
```

```python
plt.imshow(data, cmap='gray')
```

## FFT

```python
fft = np.fft.fft2(data)
fft = np.fft.fftshift(fft)
```

```python
plt.imshow(abs(np.log(fft)), cmap='gray')
```

```python
plt.imshow(np.angle(np.log(fft)), cmap='gray')
```

```python
ifft = np.fft.ifftn(fft)
```

```python
plt.imshow(abs(ifft), cmap='gray')
```
