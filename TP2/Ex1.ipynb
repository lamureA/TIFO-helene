```python
from PIL import Image
import numpy as np
import matplotlib.pylab as plt
```

## Load Image

```python
img = Image.open('../images/got.jpg')
img.load()
data = np.asarray(img)
```

```python
fig=plt.figure(figsize=(10, 10))

fig.add_subplot(1, 1, 1)
plt.imshow(img)
```

## Apply Gamma

```python
gamma = 1.3
res = np.power(img, gamma)
res = np.clip(res, 0, 255)
res = res.astype('uint8')
```

```python
fig=plt.figure(figsize=(10, 10))

fig.add_subplot(1, 1, 1)
plt.imshow(res)
```
